```python
import pandas as pd
filled_logdata_adjusted=pd.read_csv("D:/GEMimp/0915/filled_logdata_adjusted_2.csv",index_col=0)
filled_logdata_adjusted.head(5)
```

```python
zero_count = (filled_logdata_adjusted == 0).sum().sum()

total_elements = filled_logdata_adjusted.size
zero_proportion = zero_count / total_elements
zero_proportion
```

```python

from MBIMP import Impute

import pandas as pd
graph_adj=Impute.GraphBuild(input_file=filled_logdata_adjusted,k=3) 
cell_emb=Impute.trainCellEmbeddings(graph_adj)
data_imp=Impute.imputation(mbfile=filled_logdata_adjusted,embeddingfile=cell_emb,AdjGraph=graph_adj)
data_imp.head()
```

```python
import numpy as np
data_imp[data_imp == 0] = np.log10(1.01)
data_imp.to_csv('D:/GEMimp/0915/outptfile_bygemimp_2.csv', index=True) 
```
