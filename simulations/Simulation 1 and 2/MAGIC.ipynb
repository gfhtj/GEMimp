### complete datasets

```python
import magic
import scprep
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
filled_logdata_adjusted=pd.read_csv("D:/GEMimp/0915/filled_logdata_adjusted_2.csv",index_col=0)
filled_logdata_adjusted.head(5)
```

```python
newfile2=filled_logdata_adjusted
mydata2=newfile2.T
magic_op2 = magic.MAGIC()
emt_magic2 = magic_op2.fit_transform(mydata2)
fmatrix2=emt_magic2.T
ip_file2=fmatrix2
ip_file2.head(5)
```

```python
zero_count = (ip_file2 == 0).sum().sum()

total_elements = ip_file2.size
zero_proportion = zero_count / total_elements
zero_proportion
```

```python
ip_file2.to_csv('D:/GEMimp/0915/outptfile_bymagic_2.csv', index=True) 
```
