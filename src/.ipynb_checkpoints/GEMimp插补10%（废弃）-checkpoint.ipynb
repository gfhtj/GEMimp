### GEMimp插补

```python
from MBIMP import Impute
```

```python
import pandas as pd
```

```python
rawfile=pd.read_csv("D:/阅读文献同步文件夹/mbImpute code/0825new/modified_data_experiment_1.csv",index_col=0) #10%缺失
```

```python
rawfile.head(5)
```

```python
import numpy as np
rawfile = np.log(rawfile+1.01)
```

```python
rawfile.head()
```

```python
rawfile.to_csv('raw_file_bylog.csv', index=True) #原始数据
```

```python
graph_adj=Impute.GraphBuild(input_file=rawfile,k=3)
```

```python

cell_emb=Impute.trainCellEmbeddings(graph_adj)
```

```python
#Step4 scRNA-seq data imputation, the format of output file is genes x cells expression matrix.

data_imp=Impute.imputation(mbfile=rawfile,embeddingfile=cell_emb,AdjGraph=graph_adj)
```

```python
data_imp.head(5)
```

```python
data_imp.to_csv('D:/阅读文献同步文件夹/mbImpute code/0825new/output_file_byGEMimp10.csv', index=True) #利用geM插补的10%的数据，以log行书存储
```
