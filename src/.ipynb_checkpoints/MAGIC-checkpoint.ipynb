## magic 现缺失10%插补

```python
import magic
import scprep

import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
```

```python
rawfile=pd.read_csv("D:/阅读文献同步文件夹/mbImpute code/0825new/modified_data_experiment_1.csv",index_col=0) #10%缺失
```

```python
rawfile.head()
```

```python
import numpy as np
rawfile = np.log(rawfile+1.01)
```

```python
rawfile.head()
```

```python
newfile2=rawfile
mydata2=newfile2.T
magic_op2 = magic.MAGIC()
emt_magic2 = magic_op2.fit_transform(mydata2)
fmatrix2=emt_magic2.T
ip_file2=fmatrix2
```

```python
ip_file2.head(5)
```

```python
ip_file2.to_csv("D:/阅读文献同步文件夹/mbImpute code/0825new/outptfile_bymagic10.txt",sep="\t")#
```
