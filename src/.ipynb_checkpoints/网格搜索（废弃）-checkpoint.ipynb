### 以下代码是为了尝试去得到taxa abundance与zero数量的关系，结论不是很好，先搁置一下

```python
import pandas as pd
import numpy as np

# 假设 t2d 是原始数据框

# 计算每列的非零数值数量
nonzero_counts = t2d.astype(bool).sum(axis=0)

# 计算每列的零值数量
zero_counts = t2d.isna().sum(axis=0) + (t2d == 0).sum(axis=0)

# 计算每列的非零数值总和
nonzero_sums = t2d[t2d != 0].sum(axis=0)

```

```python
# 创建一个数据框以存储每列的结果
result_df = pd.DataFrame({
    'Column': t2d.columns,
    'NonZero_Count': nonzero_counts,
    'NonZero_Sum': nonzero_sums,
    'Zero_Count': zero_counts
})

# 输出结果
print("每列的非零数值数量、非零总和和零值数量：")
print(result_df)
```

```python
# 将结果转换为数据框
result_df = pd.DataFrame(result_df, columns=result_df.columns)
```

```python
result_df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# 可视化非零总和与零值数量的关系
plt.figure(figsize=(8, 5))
sns.scatterplot(data=result_df, x='Zero_Count', y='NonZero_Sum', hue='NonZero_Count', palette='viridis', size='NonZero_Count', sizes=(20, 200))
sns.regplot(data=result_df, x='Zero_Count', y='NonZero_Sum', scatter_kws={'s':50}, line_kws={'color':'red'}, order=3)
plt.title('Non-Zero Sum vs. Zero Count')
plt.xlabel('Zero Count')
plt.ylabel('Non-Zero Sum')
plt.legend(loc='best')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np

# 设置一个随机种子，以便结果可重现
np.random.seed(13)

# 假设 t2d 是原始数据框

# 定义实验比例：50%, 70%, 90%
ratios = [0.5, 0.7, 0.9]

# 计算每列的非零数值数量
nonzero_counts = t2d.astype(bool).sum(axis=0)
total_nonzero = nonzero_counts.sum()

for i, ratio in enumerate(ratios):
    # 复制原始数据以便进行不同实验
    modified_df = t2d.copy()

    # 计算要设为零的总数量
    num_values_to_zero = int(modified_df.size * ratio)

    # 计算每列要设为零的数量
    # 计算列的权重，权重越低的列分配的零值更多
    weights = 1 / (nonzero_counts + 1)  # 使用 +1 防止除以零
    column_zero_targets = (weights / weights.sum() * num_values_to_zero).astype(int)  # 每列要设为零的数量

    # 确保每列设为零的数量不超过该列的非零数值
    column_zero_targets = np.clip(column_zero_targets, 0, nonzero_counts)

    # 设为零的具体位置
    for col in modified_df.columns:
        num_zeros = column_zero_targets[col]
        if num_zeros > 0:
            nonzero_indices = modified_df[modified_df[col] != 0].index
            if len(nonzero_indices) < num_zeros:
                num_zeros = len(nonzero_indices)  # 如果非零值不足以设为零，则调整
            indices_to_zero = np.random.choice(nonzero_indices, size=num_zeros, replace=False)
            modified_df.loc[indices_to_zero, col] = 0

    # 保存修改后的数据框到文件
    modified_df.to_csv(f'D:/阅读文献同步文件夹/mbImpute code/0825new/modified_data_experiment_{i+1}.csv', index=True)

    # 输出修改后的数据集
    print(f"实验 - 总体零值占比为{ratio * 100}%：")
    print(modified_df.shape)
    print("\n" + "="*30 + "\n")

# 显示每列的非零数值大小和零值数量的关系
zero_counts = [column_zero_targets[col] for col in t2d.columns]
nonzero_and_zero_counts = pd.DataFrame({
    'NonZero_Count': nonzero_counts,
    'Zero_Count': zero_counts
})

print("每列的非零数值数量与零值数量的关系：")
print(nonzero_and_zero_counts)

```
